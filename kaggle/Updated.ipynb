```python
# IMPORTS
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

import nltk
#nltk.download('all') # only need to be downloaded once, so comment out once ran
import re
from nltk.corpus import stopwords

descriptors = ['headline', 'article', 'category']

# Load in all CSVs into pandas dataframes
df1 = pd.read_csv('inshort_news_data-1.csv', skiprows = 1, names = descriptors, delimiter = ',')
df2 = pd.read_csv('inshort_news_data-2.csv', skiprows = 1, names = descriptors, delimiter = ',')
df3 = pd.read_csv('inshort_news_data-3.csv', skiprows = 1, names = descriptors, delimiter = ',')
df4 = pd.read_csv('inshort_news_data-4.csv', skiprows = 1, names = descriptors, delimiter = ',')
df5 = pd.read_csv('inshort_news_data-5.csv', skiprows = 1, names = descriptors, delimiter = ',')
df6 = pd.read_csv('inshort_news_data-6.csv', skiprows = 1, names = descriptors, delimiter = ',')
df7 = pd.read_csv('inshort_news_data-7.csv', skiprows = 1, names = descriptors, delimiter = ',')

df3.head() # Verify df loaded correctly 

df1['category'].unique() # print unique categories

df = pd.concat([df1,df2,df3,df4,df5,df6,df7]) # Vertically concatenate all dfs into one

# Shuffle dataframe
df = df.sample(frac=1).reset_index(drop=True)

set(df['category']) # Set of unique labels for classification

def remove_non_letters(word):
    # TODO: Modify in future to remove numbers? Note: this will turn 3D/4D --> D
    new_word = ''
    for letter in word:
        if letter.isalpha() == True or letter.isnumeric() == True:
            new_word += letter
    
    return new_word

def clean_article(article):
    # Split article into individual words/tokens
    tokens = article.split(' ')
    
    # Make each word lower-case
    tokens = [t.lower() for t in tokens]
    
    # Remove special/non-alphabetic/numeric characters from each word
    tokens = [remove_non_letters(t) for t in tokens]
    
    # Remove numbers/numeric tokens from the list
    tokens = [t for t in tokens if t.isnumeric() == False]
    
    tokens = [w for w in tokens if w not in stopwords.words('english')]
    
    # TODO: Look into this, because some of the changes are not perfect (has --> ha)
    # Lemmatize words (studying/studies/studied --> study)
    # lemmatizer = WordNetLemmatizer()
    # tokens = [lemmatizer.lemmatize(t) for t in tokens]
    
    # Stem all tokens
    ps = PorterStemmer()
    tokens = [ps.stem(t) for t in tokens]
    
    return ' '.join(tokens)

X = df['article'].values # [FIRST] Use article as the only feature
labels = df['category'].values # Category as the label

X = [clean_article(a) for a in X] # pre-process each article in X

#Converting text into numerical format for bag of words model
vectorizer = CountVectorizer() # Use max_features?
X_vectorized = vectorizer.fit_transform(X)

# Change y into numeric labels
mapping = {'automobile': 0, 'entertainment': 1, 'politics': 2, 'science': 3, 'sports': 4, 'technology': 5, 'world': 6}
y = []
for label in labels:
    y.append(mapping[label])
    
y = np.array(y)

X_vectorized

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(30, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(30, activation = tf.nn.relu))
#10 digits
model.add(tf.keras.layers.Dense(7, activation = tf.nn.softmax))

#Could change adam
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'] )

k = 5

# Split array into 6 subsets, only run 5-fold CV on first 5 subsets, use fold #6 as external validation for testing
x_folds = np.array_split(X_vectorized.toarray(), k)
y_folds = np.array_split(y, k)

val_accuracies = []

for i in range(k):
    
   # Create a copy of x_fold and y_folds
    x_folds_copy = x_folds.copy()
    y_folds_copy = y_folds.copy()


    # Access current fold as testing subset -- held out testing set
    X_test_folds = np.array(x_folds[i])
    y_test_folds = np.array(y_folds[i])

    # Training subset is everything *except* current fold
    x_folds_copy.pop(i)
    y_folds_copy.pop(i)
    
    x_nn_train = np.concatenate(x_folds_copy)
    y_nn_train = np.concatenate(y_folds_copy)

    m_nn = model.fit(x_nn_train, y_nn_train, validation_split=0.2, epochs = 10)
    
    val_accuracy = pd.DataFrame(m_nn.history)['val_accuracy'][9] # accuracy after 10 epochs for the current fold
    val_accuracies.append(val_accuracy)
    
    print("Fold " + str(i + 1) + " validation accuracy: " + str(val_accuracy))
    
    #View training and validation curves
    pd.DataFrame(m_nn.history).plot(figsize=(8, 5))
    plt.grid(True)
    plt.gca().set_ylim(0, 1.5) # set the vertical range to [0-1.5]
    plt.show()
    
print("Mean validation accuracy over 5 folds:", (sum(val_accuracies) / k) * 100)
    
```
